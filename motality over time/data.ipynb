```python
import plotly.express as px
import pandas as pd
import numpy as np
```

```python
path = 'https://github.com/nytimes/covid-19-data/blob/1095bf5cee95d2c3a94672b4fc0ea72a2b87c324/excess-deaths/deaths.csv?raw=true'
df = pd.read_csv(path,index_col=None)
df = df[df["excess_deaths"].isna() == False]
df['P-score'] = df["excess_deaths"]/df["expected_deaths"]
df['month'] = df['month'].apply(str)
```

```python
df.loc[653]['start_date']
```

```python
df['start_date'] = np.where(df['frequency'] == 'monthly', "2020-"+df['month']+"-01",df['start_date'])
```

```python
countries = df['country']
countries = countries.unique()
countries_to_hide = np.delete(countries, 0)
```

```python
df
```

```python
def update_fig(countries=['Austria', 'Brazil']):
    fig = px.line(df, x="start_date", y="P-score", color='country',title='Excess Mortality by Country')
    fig.layout.yaxis.tickformat = ',.0%'
    fig.update_layout(
        yaxis_range=[-0.2,3],
        hovermode= 'x unified')
    fig.for_each_trace(lambda trace: trace.update(visible='legendonly')
                      if trace.name in countries_to_hide else ())
    fig.show()
```

```python
update_fig()
```
